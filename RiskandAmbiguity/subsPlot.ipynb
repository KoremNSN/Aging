```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
db = pd.read_csv('mon.csv')
subs = db['sub'].unique()
```

```python
for participant in subs:
    print(participant)
    df = db[db['sub'] == participant]
    df_ambg0 = df[df['ambiguity']==0]
    df_ambgN0 = df[df['ambiguity']>0]

    sns.lmplot(x="value", y="choice", data=df_ambg0, hue= 'risk',
               logistic=True, y_jitter=.1)
    sns.lmplot(x="value", y="choice", data=df_ambgN0, hue= 'ambiguity',
               logistic=True, y_jitter=.1)
    plt.show()
    
```

```python
df_ambg0 = db[db['ambiguity']==0]
ax = sns.lmplot(x="value", y="choice", data=df_ambg0, hue= 'risk',
                 logistic=True, x_jitter=.5, y_jitter=.1)
```

```python
df_ambgN0 = db[db['ambiguity']>0]
sns.lmplot(x="value", y="choice", data=df_ambgN0, hue= 'ambiguity',
                 logistic=True, x_jitter=.5, y_jitter=.3)
```
